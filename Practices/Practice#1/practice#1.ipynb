## 2. Matrix Operations

Two common matrix operations are **transpose** and **concatenation**.

**Transpose** flips a matrix over its main diagonal — element at row $i$, column $j$ moves to row $j$, column $i$:
$$A^T_{ij} = A_{ji}$$
In numpy: `A.T` or `np.transpose(A)`.

**Concatenation** joins matrices along a given axis:

| Function | Axis | Requirement | Representation |
|----------|------|-------------|----------------|
| `np.hstack([A, B])` | columns (axis=1) | same number of **rows** | $\begin{bmatrix}A&B \end{bmatrix}$ |
| `np.vstack([A, B])` | rows (axis=0) | same number of **columns** | $\begin{bmatrix}A\\B\end{bmatrix}$  |
| `np.concatenate([A, B], axis=k)` | axis k | general form |

````python
import numpy as np

# Transpose
ex = np.array([[1, 2, 3],
               [4, 5, 6]])  # shape (2, 3)
print('ex:\n', ex)
print('ex.T (shape 3x2):\n', ex.T)

# Horizontal stack (hstack): same row count
ex_a = np.array([[1, 2], [3, 4]])   # shape (2, 2)
ex_b = np.array([[5, 6], [7, 8]])   # shape (2, 2)
print('hstack [ex_a | ex_b]:\n', np.hstack([ex_a, ex_b]))   # shape (2, 4)

# Vertical stack (vstack): same column count
ex_c = np.array([[9, 10]])          # shape (1, 2)
print('vstack [ex_a / ex_c]:\n', np.vstack([ex_a, ex_c]))   # shape (3, 2)
````

## 3. Matrix Multiply

There are three distinct "multiplication" operations on matrices:

| Operation | Symbol | Numpy | Description |
|-----------|--------|-------|-------------|
| Scalar multiply | $k \cdot A$ | `k * A` | Every element multiplied by scalar $k$ |
| Element-wise (Hadamard) | $A \odot B$ | `A * B` | Multiply corresponding elements (shapes must match) |
| Matrix multiply | $AB$ | `A @ B` | Standard linear algebra multiplication |

For matrix multiplication $A @ B$:
if $A$ is $(m \times n)$ and $B$ is $(n \times k)$, the result is $(m \times k)$.
The **inner dimensions must match**.


````python
import numpy as np

A_ex = np.array([[1, 2],
                 [3, 4]], dtype=float)
B_ex = np.array([[5, 6],
                 [7, 8]], dtype=float)

# Scalar multiplication
print('3 * A_ex:\n', 3 * A_ex)

# Element-wise multiplication (Hadamard product)
print('A_ex * B_ex (element-wise):\n', A_ex * B_ex)

# Matrix multiplication
print('A_ex @ B_ex (matrix multiply):\n', A_ex @ B_ex)

# Note: (A_ex @ B_ex)[0,0] = 1*5 + 2*7 = 19
````

## 4. Inverse Matrix

The **inverse** of a square matrix $A$, denoted $A^{-1}$, satisfies:
$$A A^{-1} = A^{-1} A = I$$

In numpy: `np.linalg.inv(A)` — $A$ must be **square** and **non-singular** ($\det(A) \neq 0$).

The **Moore-Penrose pseudoinverse** $A^+$ generalizes the inverse to any matrix (rectangular or singular).
For a full column-rank matrix it equals $(A^T A)^{-1} A^T$, and in general:
$$A^+ = V \Sigma^+ U^T$$
where $V$, $\Sigma$, and $U$ come from SVD (Singular Value Decomposition) operation. We will talk about SVD later. 

In numpy: `np.linalg.pinv(A)` — works for **any shape**.

> **Key property**: $A^+$ gives the minimum-norm least-squares solution to $Ax = b$ via $x = A^+ b$.


````python
import numpy as np

# Regular inverse (square, non-singular matrix)
M_ex = np.array([[1, 2],
                 [3, 4]], dtype=float)
M_inv_ex = np.linalg.inv(M_ex)
print('M_ex @ inv(M_ex) (should be I):\n', M_ex @ M_inv_ex)

# Pseudoinverse — works for non-square matrices
P_ex = np.array([[1, 2, 3],
                 [4, 5, 6]], dtype=float)  # shape (2, 3)
P_pinv_ex = np.linalg.pinv(P_ex)           # shape (3, 2)
print('P_ex shape:', P_ex.shape, '  pinv shape:', P_pinv_ex.shape)
print('P_ex @ pinv(P_ex) (should be ~I_2x2):\n', P_ex @ P_pinv_ex)
````

## 5. Singular Value Decomposition (SVD)

SVD decomposes any $m \times n$ matrix $A$ into:
$$A = U \Sigma V^T$$

| Factor | Shape | Description |
|--------|-------|-------------|
| $U$ | $m \times m$ | Left singular vectors (orthonormal columns) |
| $\Sigma$ | $m \times n$ | Diagonal matrix of singular values $\sigma_1 \geq \sigma_2 \geq \ldots \geq 0$ |
| $V^T$ | $n \times n$ | Right singular vectors (orthonormal rows) |

In numpy: `U, S, Vt = np.linalg.svd(A, full_matrices=False)` — economy (thin) SVD;
`S` is a 1-D array of singular values, not a 2-D diagonal matrix.

### Solving $Ax = b$ with SVD (Least-Squares)

When $A$ is **overdetermined** (more equations than unknowns), $Ax = b$ generally has no exact solution.
SVD gives the **least-squares solution** minimizing $\|Ax - b\|_2^2$:

$$x = V \Sigma^+ U^T b$$

where $\Sigma^+$ replaces each $\sigma_i$ with $1/\sigma_i$.

Equivalent approaches in numpy:
```python
x = Vt.T @ np.diag(1/S) @ U.T @ b          # explicit SVD formula
x = np.linalg.pinv(A) @ b                   # pseudoinverse
x, _, _, _ = np.linalg.lstsq(A, b, rcond=None)  # built-in least-squares
```


````python
import numpy as np

# SVD decomposition example
A_ex = np.array([[1, 2],
                 [3, 4],
                 [5, 6]], dtype=float)  # (3x2) overdetermined

U, S, Vt = np.linalg.svd(A_ex, full_matrices=False)  # economy SVD
print('U shape:', U.shape, '  S:', S.round(4), '  Vt shape:', Vt.shape)
print('Reconstruction U @ diag(S) @ Vt:\n', (U @ np.diag(S) @ Vt).round(10))

# Solve Ax = b (least-squares) — three equivalent methods
b_ex = np.array([1.0, 2.0, 3.0])

x_svd_ex  = Vt.T @ np.diag(1 / S) @ U.T @ b_ex   # SVD formula
x_pinv_ex = np.linalg.pinv(A_ex) @ b_ex            # pseudoinverse
x_lst_ex, _, _, _ = np.linalg.lstsq(A_ex, b_ex, rcond=None)  # lstsq

print('x (SVD formula):', x_svd_ex.round(6))
print('x (pinv):       ', x_pinv_ex.round(6))
print('x (lstsq):      ', x_lst_ex.round(6))
print('Residual ||Ax-b||:', np.linalg.norm(A_ex @ x_svd_ex - b_ex).round(6))
````
